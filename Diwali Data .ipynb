```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
df=pd.read_csv("Diwali Sales Data.csv",encoding='unicode_escape')
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values('Amount')
```

```python
sns.barplot(x='Gender',y='Amount',data=sales)
```

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
df.columns
```

```python
c=df.groupby(["State"],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=c,x='State',y='Orders')
```

```python
c=df.groupby(["State"],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=c,x='State',y='Amount')
```
